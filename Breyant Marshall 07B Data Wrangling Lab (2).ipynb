# Section 2: Missing Data
<hr style="border-top:1px dashed">

<b>How to deal with missing data?</b>

<ol>
    <li>drop data<br>
        a. drop the whole row<br>
        b. drop the whole column
    </li>
    <li>replace data<br>
        a. replace it by mean<br>
        b. replace it by frequency<br>
        c. replace it based on other functions
    </li>
</ol>

Whole columns should be dropped only if most entries in the column are empty. In our dataset, none of the columns are empty enough to drop entirely.
We have some freedom in choosing which method to replace data; however, some methods may seem more reasonable than others. We will apply each method to many different columns:

# Task 3: What do you see that needs to be .replaced()?

```python
# Hint: What character is being used instead of null?
# Answer: ?
```

# Task 4: Replace the null character in the dataframe with np.nan

```python
#Hint: Inplace = True
df.replace("?", np.nan, inplace = True)
df
```

# Task 5-9: Replace these missing values with their column's mean:

    "normalized-losses": 41 missing data
    "stroke": 4 missing data
    "bore": 4 missing data
    "horsepower": 2 missing data
    "peak-rpm": 2 missing data
For Each Column:
        Calculate the average of the column
        Set the next column to a temporary variable
        Replace "NaN" by mean value in the column using the variable
        Inplace = True when you are confident

```python
# normalized-losses
avg_nloss = df["normalized-losses"].astype("float").mean(axis=0)
avg_nloss
```

```python
df["normalized-losses"].replace(np.nan, avg_nloss, inplace=True)
df
```

```python
# stroke
avg_stroke=df['stroke'].astype('float').mean(axis=0)
avg_stroke
```

```python
df["stroke"].replace(np.nan, avg_stroke, inplace=True)
df
```

```python
# bore
avg_bore=df['bore'].astype('float').mean(axis=0)
avg_bore
```

```python
df["bore"].replace(np.nan, avg_bore, inplace=True)
df
```

```python
# horsepower
avg_hp = df['horsepower'].astype('float').mean(axis=0)
avg_hp
```

```python
df['horsepower'].replace(np.nan, avg_hp, inplace=True)
df
```

```python
# peak-rpm
avg_peak_rpm=df['peak-rpm'].astype('float').mean(axis=0)
avg_peak_rpm
```

```python
df['peak-rpm'].replace(np.nan, avg_peak_rpm, inplace=True)
df
```

# Task 11: Drop the whole row of data:

# Task 13: Use a function to list the data types for each column:

```python
df.dtypes
```

<p>As we can see above, some columns are not of the correct data type. Numerical variables should have type 'float' or 'int', and variables with strings such as categories should have type 'object'. For example, 'bore' and 'stroke' variables are numerical values that describe the engines, so we should expect them to be of the type 'float' or 'int'; however, they are shown as type 'object'. We have to convert data types into a proper format for each column using the "astype()" method.</p> 


# Task 14-19: Covert Column's Data Type to correct Data Type:
    float - Bore, Stroke, Price
    int64 - normalized-losses, horsepower, peak-rpm

```python
# hint: Use the astype function: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.astype.html
```

```python
df[["bore", "stroke", "price"]] = df[["bore", "stroke", "price"]].astype("float")
```

```python
df[["normalized-losses", "horsepower", "peak-rpm"]] = df[["normalized-losses","horsepower", "peak-rpm"]].astype("int64")
```
